```python
import mysql.connector

conn = mysql.connector.connect(user='root',password='root',host='localhost')

cursor = conn.cursor()

cursor.execute("CREATE DATABASE if NOT EXISTS housingdata")

cursor.execute("USE housingdata")

cursor.execute('''
CREATE TABLE IF NOT EXISTS property_details (
    id varchar(50) NOT NULL,
    start_date DATE NOT NULL,
    end_date DATE NOT NULL,
     created_on DATE NOT NULL,
     lat INT NOT NULL,
     lon INT NOT NULL,
     l1 VARCHAR(20) NOT NULL,
     l2 VARCHAR(30) NOT NULL,
     l3 VARCHAR(25) NOT NULL,
     rooms INT NOT NULL,
     bathrooms INT NOT NULL,
     bedrooms INT NOT NULL,
     surface_total INT NOT NULL,
     surface_covered INT NOT NULL,
     PRIMARY KEY(id)
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS property_price_details (
    id varchar(50) NOT NULL,
    price INT NOT NULL,
    currency VARCHAR(7) NOT NULL,
    price_period VARCHAR(15) NOT NULL,
    property_type VARCHAR(25) NOT NULL,
    operation_type VARCHAR(15) NOT NULL,
    PRIMARY KEY(id)
)
''')
```

```python
import mysql.connector

from sqlalchemy import create_engine

eng = create_engine('mysql+mysqlconnector://root:root@localhost/housingdata')
property_details.to_sql('property_details', con = eng, if_exists='replace', index=False)
property_price_details.to_sql('property_price_details', con = eng, if_exists='replace', index=False)
```

```python
# 1) Retrieve properties that have a price greater than 1 million and are located in "Estados Unidos" (l1).

sql_query = 'select * from property_details pd where l1 = "Estados Unidos" and id in (select id from property_price_details where price>1000000)'
query1 = pd.read_sql(sql_query,con=conn)
query1
```

```python

# 2) Categorize properties based on their surface area as 'Small' if it's less than 50 square meters, 'Medium' if it's between 50 and 100 square meters, and 'Large' if it's greater than 100 square meters:

sql_query = "select id, case when surface_total<50 then 'Small' when surface_total between 50 and 100 then 'Medium' else 'Large' end as surface_area_category from property_details"
query2 = pd.read_sql(sql_query,con=conn)
query2
```

```python
# 3) List all properties (id) in the "Belgrano" neighborhood (l3) that have the same number of bedrooms and
# bathrooms as another property in the dataset:

sql_query = '''Select Distinct pr1.id, pr1.bedrooms, pr1.bathrooms
from property_details pr1
join property_details pr2 on  pr1.bedrooms = pr2.bedrooms and pr1.bathrooms = pr2.bathrooms
where pr1.l3 = 'Belgrano' and pr1.id <> pr2.id'''
query3 = pd.read_sql(sql_query,con=conn)
query3
```

```python
# 4) Calculate the average price per square meter (price / surface_total) for each property type (property_type) in
# the "Belgrano" neighborhood (l3):

sql_query = "select property_type,round(avg(price/surface_total),2) as Average_price_belgrano from property_details pd join property_price_details as ppd using(id) where pd.l3 = 'Belgrano'  group by property_type"
query4 = pd.read_sql(sql_query,con=conn)
query4
```

```python
# 5) Identify properties that have a higher price than the average price of properties with the same number of
# bedrooms and bathrooms.

conn = mysql.connector.connect(user='root', password='root', host='localhost', database='housingdata')
sql_query = "select id,price,round(avg(price),2) as average_price from property_details as pd join property_price_details ppd using(id) where bedrooms=bathrooms group by id,price having price>average_price"
query5 = pd.read_sql(sql_query,con=conn)
query5

```

```python
# 6) Calculate the cumulative price for each property type, ordered by the creation date.

sql_query = "SELECT property_type, SUM(price) AS cumulative_price, MIN(created_on) AS min_created_on FROM property_price_details ppd JOIN property_details pd USING (id) GROUP BY property_type ORDER BY min_created_on"
query6 = pd.read_sql(sql_query,con=conn)
query6



```

```python
# 7) Identify the 10 locations (l3) with the highest total surface area (sum of surface_total) of properties listed for
# sale (operation_type = 'Venta'):

sql_query = '''select * from (select l3,rank() over(order by total_surface_area) as top
              from (select l3,sum(surface_total) as total_surface_area from property_details pd join property_price_details ppd using(id)
              where operation_type='Venta' group by l3) as t1) as t2 where top<=10'''
query7 = pd.read_sql(sql_query,con=conn)
query7


# can also use dense_rank

```

```python
# 8) Find the top 5 most expensive properties (based on price) in the "Palermo" neighborhood (l3) that were listed
# in August 2020:

sql_query = '''select * from(
select id,dense_rank() over(order by price desc) as expense_order,created_on
from property_details as pd
join property_price_details as ppd
using(id)
where l3='Palermo' and month(created_on) = 8 and year(created_on)=2020) as t
where expense_order<=5'''
query8 = pd.read_sql(sql_query,con=conn)
query8


```

```python
# 9) Find the top 3 properties with the highest price per square meter (price divided by surface area) within each
# property type.

sql_query = "select * from (select property_type,id, dense_rank() over(partition by property_type order by (price/surface_total) desc) as top3 from property_details pd join property_price_details ppd using(id)) as t where top3<=3"
query9 = pd.read_sql(sql_query,con=conn)
query9


```

```python
# 10) Find the top 3 locations (l1, l2, l3) with the highest average price per square meter (price / surface_total) for
# properties listed for sale (operation_type = 'Venta') in the year 2020. Exclude locations with fewer than 10
# properties listed for sale in 2020 from the results.

sql_query = "select * from (select *,dense_rank() over(order by average_price desc) as top3 from (select l1,l2,l3,round(avg(price/surface_total),2) as average_price from property_details pd join property_price_details ppd using(id) where operation_type='Venta' and year(start_date)=2020 and l3 is not null group by l1,l2,l3 having count(id)>10) as t1) as t2 where top3<=3"
query10 = pd.read_sql(sql_query,con=conn)
query10


```
